```python
#bin the data
winsize = 10.0
dtovr = winsize*0.5
comment = 'ForJSDivergence'
segment_info = [0,0]

bdfs = {}
birds = ['B1083', 'B1235', 'B1075']
for bird in birds:
    block_path = bps[bird]
    bfdict = tp2.dag_bin(block_path, winsize, segment_info, cluster_group=['Good'], dt_overlap=dtovr, comment=comment)
    bdf = glob.glob(os.path.join(bfdict['raw'], '*.binned'))[0]
    print(bdf)
    bdfs[bird] = bdf
```

```python
bird = 'B1083'
bdf = bdfs[bird]
f = h5py.File(bdf, 'r')
stims = list(f.keys())
nclus = f.attrs['nclus']
cells = np.random.permutation(np.arange(nclus))
popAcells = cells[:int(round(nclus/2))]
popBcells = cells[int(round(nclus/2)):]
pop_A = {}
pop_B = {}
for stim in stims:
    poptens = np.array(f[stim]['pop_tens'])
    pop_A[stim] = poptens[popAcells, :, :]
    pop_B[stim] = poptens[popBcells, :, :]
```

```python
scg_popA={}
scg_popB={}
thresh = 6.0
for stim in stims:
    print(stim)
    scg_trialsA ={}
    scg_trialsB = {}
    pop_A_stim = pop_A[stim]
    pop_B_stim = pop_B[stim]
    ncells, nwin, ntrials = np.shape(pop_A_stim)
    for trial in range(ntrials):
        scg_trialsA[trial] = sa.computeChainGroup(pop_A_stim, thresh, trial)
        scg_trialsB[trial] = sa.computeChainGroup(pop_B_stim, thresh, trial)
    scg_popA[stim] = scg_trialsA
    scg_popB[stim] = scg_trialsB
```
