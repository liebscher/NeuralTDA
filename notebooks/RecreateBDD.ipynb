```python
# fig 2

# generate a random graph:
N = 200
def ErdosRenyi(plink, N):
    adj = np.random.rand(N, N)
    adj = (adj + adj.T)/2.0
    adjData = (adj < plink).astype(int)
    adjData = adjData - np.diag(np.diag(adjData))
    Edata = sc.simplicialChainGroups(sc.adjacency2maxsimp(adjData, range(N)))
    Ldata = sc.compute_laplacian(Edata, 0)

    Ddata = np.diag(np.sum(adjData, axis=0))
    Ldata_graph = Ddata - adjData
    return (adjData, Ldata)
```

```python
invbetas = np.linspace(0.01, 1000, 500)
plinks = np.linspace(0, 1, 5)

res = np.zeros((len(invbetas), len(plinks)))

N=200
for i, plink in enumerate(plinks):
    (A, L) = ErdosRenyi(plink, N)
    for j, beta in enumerate(invbetas):
        print((i,j))
        rho = sc.densityMatrix(L, -1/beta)
        S = sc.Entropy(rho)
        res[j, i] = S
```

```python
res = res / (np.log(N)/np.log(2))

plt.semilogx(invbetas, res)
```

```python
invbetas = np.linspace(0.01, 1000, 500)
plinks = np.linspace(0, 1, 5)

res = np.zeros((len(invbetas), len(plinks)))

N=200
for i, plink in enumerate(plinks):
    (A, L) = ErdosRenyi(plink, N)
    L = np.diag(np.sum(A, axis=0)) - A
    (v, w) = np.linalg.eig(L)
    v = np.real(v)
    for j, beta in enumerate(invbetas):
        Z = np.sum(np.exp(-v/beta))
        lambdarho = (1/Z)*np.exp(-v/beta)
        S=0
        #for l in lambdarho:
        #    if l < 1e-15:
        #        S += 0
        #    else:
        #        S += -l*np.log(l)/np.log(2.0)
        S = -np.sum(np.multiply(lambdarho, np.log(lambdarho)/np.log(2.0)))
        res[j, i] = S
```

```python
res = res / (np.log(N)/np.log(2))

plt.semilogx(invbetas, res)
```
